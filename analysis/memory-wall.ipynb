```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python

years = np.arange(1980, 2007)
relative_years = years - 1980

cpu_performance = 1.60 ** relative_years
dram_performance = 1.07 ** relative_years

plt.style.use('default')
fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(years, cpu_performance, marker='o', linestyle='-', color='#ADD8E6', label='CPU Performance')
ax.plot(years, dram_performance, marker='o', linestyle='-', color='#90EE90', label='DRAM Performance')

ax.set_yscale('log')

ax.set_xlabel('Year', fontsize=14, color='dimgray')
ax.set_ylabel('Relative Performance (Log Scale)', fontsize=14, color='dimgray')
ax.set_title('CPU vs. DRAM Performance Evolution (1980-2006)', fontsize=16, color='black')

ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True, nbins=13))
ax.tick_params(axis='x', labelrotation=45, colors='dimgray')
ax.tick_params(axis='y', colors='dimgray')

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
ax.set_yticks([1, 10, 100, 1000, 10000])
ax.get_yaxis().set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ',')))

ax.grid(True, which="both", ls="--", color='lightgray', alpha=0.7)

fig.patch.set_facecolor('white')
ax.set_facecolor('whitesmoke')

ax.text(1982, 200, "Moore's Law", fontsize=18, fontweight='bold', color='gray')

ax.annotate(
    'CPU growing 60%/year',
    xy=(2006, cpu_performance[-1]),
    xytext=(1990, 8000),
    fontsize=14,
    color='darkblue',
)

ax.annotate(
    'DRAM grows 7%/year',
    xy=(2006, dram_performance[-1]),
    xytext=(1995, 0.8),
    fontsize=14,
    color='darkgreen',
)

ax.annotate(
    '',
    xy=(2005, dram_performance[-2]),
    xytext=(2005, cpu_performance[-2]),
    arrowprops=dict(arrowstyle='<->', color='darkslategray', lw=3)
)

y_midpoint = np.sqrt(dram_performance[-2] * cpu_performance[-2])
ax.text(
    2005.5, y_midpoint, "The Memory Wall",
    fontsize=12,
    color='darkslategray',
    rotation=90,
    va='center',
    ha='left'
)

ax.text(
    1991, 100,
    'Processor-memory\nperformance gap\ngrows ~50%/year',
    fontsize=14,
    color='darkslategray',
    fontweight='bold',
    ha='center'
)

ax.legend(fontsize=12, loc='upper left', frameon=True, facecolor='white', edgecolor='lightgray')

plt.tight_layout()

filename = 'cpu_dram_evolution.png'
plt.savefig(filename, dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
print(f"Chart saved as {filename}")

plt.show()
```
